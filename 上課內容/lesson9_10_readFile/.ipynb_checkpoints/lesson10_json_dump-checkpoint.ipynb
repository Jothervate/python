```python
import csv
with open('個股日成交資訊.csv','r',encoding='utf-8') as file:
    next(file) #第一個row不要
    rows=csv.reader(file)
    #過濾資料
    stocks = [{'證券名稱':row[1],'開盤價':row[4],'最高價':row[5],'最低價':row[6],'收盤價':row[7]} for row in rows]    

print(stocks)
```

```python
#寫出
import json
with open('stock.json','w',encoding='utf-8') as file:
    json.dump(stocks,file)
print("寫入json成功") 
```
