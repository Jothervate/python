```python
import numpy as np
import pandas as pd

scores=np.random.randint(50, 101, size=(50,5))
scores_pd = pd.DataFrame(scores,columns=['國文', '英文', '數學', '地理', '歷史'],index=range(1,51))
scores_pd.index.name = "學號"
scores_pd.columns.name = "科目"
scores_sum=scores_pd.sum(axis=1)
scores_mean = scores_pd.mean(axis=1)
scores_pd['總分'] = scores_sum
scores_pd['平均'] = scores_mean
scores_pd['名次'] = scores_sum.rank(method='first',ascending=False).astype(int)
scores_pd.sort_values(by='名次')
```
