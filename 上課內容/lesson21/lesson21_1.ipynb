```python
%matplotlib notebook
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

font = FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=8) #forMac
#font = FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8) #forWindow
```

```python
xlsx = pd.ExcelFile('總統.xls')
all=pd.read_excel(xlsx,'中　央')
all
```

```python
df=all.iloc[5:,:4]
df.columns = ['縣市','朱立倫','蔡英文','宋楚瑜']
df.name='第14任總統、副總統選舉候選人得票數一覽表'
df
```

```python
#更改欄位資料類型
df.dtypes
df['朱立倫']=df['朱立倫'].astype(np.int64)
df['蔡英文']=df['蔡英文'].astype(np.int64)
df['宋楚瑜']=df['宋楚瑜'].astype(np.int64)
df.dtypes
```

```python
df=df.reset_index(drop=True)
df
```

```python
df = df.set_index('縣市')
df
```

```python
df.columns.name = '總統候選人'
df
```

```python
dfSum = df.sum()
dfSum
```

```python

figure1,ax1=plt.subplots()
dfSum.plot(ax=ax1,kind='bar',color=['b','g','orange'])
ax1.set_xticklabels(dfSum.index,fontproperties = font,rotation=30)
ax1.set_xlabel('總統候選人',fontproperties = font)
ax1.set_yticklabels(['0','1佰萬','2佰萬','3佰萬','4佰萬','5佰萬','6佰萬','7佰萬'],fontproperties = font)
for p in ax1.patches:
    width, height= p.get_width() , p.get_height()
    x,y = p.get_xy()
    ax1.annotate(f'{height}票',(x+width/5,y+height+100000),fontproperties=font)
```
