```python
import numpy as np

scores_data = np.random.randint(50,101,size=(50,5))
scores_data
```

```python
import pandas as pd
scores_frame=pd.DataFrame(scores_data,columns=['國文','英文','數學','地理','歷史'],index=range(1,51))
scores_frame.columns.name='科目'
scores_frame.index.name = '學號'
scores_frame
```

```python
sum_series=scores_frame['國文'] + scores_frame['英文'] + scores_frame['數學']  + scores_frame['地理'] + scores_frame['歷史']
```

```python
scores_frame["總分"] = sum_series
scores_frame["平均"] = scores_frame["總分"] / 5.0
scores_frame
```
