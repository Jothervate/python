```python
import numpy as np
import math
import matplotlib.pyplot as plt

t = np.arange(0,2.5,0.1)
y1 = np.sin(t * math.pi)
y2 = np.sin(t * math.pi + math.pi/2)
y3 = np.sin(t * math.pi - math.pi/2)
plt.figure(figsize=(8,5))
plt.plot(t,y1)
plt.plot(t,y2)
plt.plot(t,y3)
plt.show()
```

```python
t = np.arange(0, 5, 0.1)
y1 = np.sin(math.pi * t)
y2 = np.sin(math.pi * t + math.pi/2)
plt.figure(figsize=(8,5))
plt.subplot(211)
plt.plot(t,y1, 'b-.')
plt.subplot(212)
plt.plot(t,y2, 'r--')
plt.show()
```
