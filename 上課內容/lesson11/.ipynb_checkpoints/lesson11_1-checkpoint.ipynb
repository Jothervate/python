```python
import matplotlib.pyplot as plt
x = ['chinese', 'english', 'math', 'history', 'discover']
y1 = [1, 2, 5, 6, 7]
y2 = [4, 5, 6, 7, 8]
plt.figure(figsize=(8,5))
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x = ['chinese', 'english', 'math', 'history', 'discover']
y1 = [1, 2, 5, 6, 7]
y2 = [4, 5, 6, 7, 8]
plt.figure(figsize=(8,5))
plt.bar(x,y1,align="edge",width=0.2,color='b')
plt.bar(x,y2,align="edge",width=-0.2,color='g')
plt.show()
```

```python
#使用list的做法，建立總分
import random
chinese = [random.randint(60,100) for _ in range(50)]
english = [random.randint(60,100) for _ in range(50)]
print(f'chinese={chinese}')
print(f'english={english}')
```

```python
result = [chinese[i]+english[i] for i in range(50)]
print(f'總分={result}')
```

```python
import numpy as np
ndChinese=np.array(chinese)
ndEnglish = np.array(english)
ndResult=ndChinese + ndEnglish
```

```python
ndResult
```

```python
xRange = range(50)
plt.figure(figsize=(8,5))
plt.plot(xRange,ndResult)
plt.show()
```
