```python
import matplotlib.pyplot as plt
x = ['chinese', 'english', 'math', 'history', 'discover']
y1 = [1, 2, 5, 6, 7]
y2 = [4, 5, 6, 7, 8]
plt.figure(figsize=(8,5))
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
x = ['chinese', 'english', 'math', 'history', 'discover']

y2 = [4, 5, 6, 7, 8]
plt.figure(figsize=(8,5))
plt.bar(x,y1,align="edge",width=0.35,color='b')
plt.bar(x,y2,align="edge",width=-0.35,color='g')
plt.show()
```
