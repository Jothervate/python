```python
import numpy as np
import pandas as pd

dataFrame = pd.read_csv('各鄉鎮市區人口密度.csv')
dataFrame.columns = dataFrame.loc[0]
dataFrame.drop(index=0,inplace=True)
dataFrame = dataFrame[:-5]
dataFrame
```

```python
def a(x):
    if len(x) > 4:
        return x[:3]
    else:
        return x
    
    
dataFrame['區域別'].map(a)
```

```python
state = dataFrame['區域別'].map(lambda x:x[:3] if len(x) > 4 else x)
dataFrame['縣市'] = state
dataFrame
```

```python
areas=dataFrame['區域別'].map(lambda x:x[-3:] if len(x)>4 else x)
dataFrame['區域別'] = areas
dataFrame=dataFrame.reindex(columns=['縣市','年底人口數','土地面積'])
```

```python
dataFrame.dtypes
```

```python
#dataFrame['年底人口數'].astype(np.int32)
def toInt(x):
    try:
        return int(x)
    except:
        return 0
dataFrame['年底人口數'] = dataFrame['年底人口數'].map(toInt)
dataFrame['土地面積'] = dataFrame['土地面積'].astype(np.float32)
```

```python
def cal(s):
    return s.sum()
dataFrame[['年底人口數','土地面積']].apply(cal)
```

```python
def cal(s):
    return pd.Series([s.sum(),s.max(),s.min()],index=['加總','最大','最小'])
dataFrame[['年底人口數','土地面積']].apply(cal)
```
