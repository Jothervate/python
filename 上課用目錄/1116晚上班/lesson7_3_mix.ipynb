```python
#讀取csv格式檔案
import csv

with open('歷年國內主要觀光遊憩據點遊客人數月別統計.csv',encoding="utf-8",newline="") as file:
    csv_reader = csv.reader(file)
    all_list = list(csv_reader)
```

```python
taipei_2021 = []
for item in all_list:
    if item[4] == '臺北市' and item[0] == '2021':
        taipei_2021.append(item)
print(taipei_2021)
```

```python
taichung_2021 = []
for item in all_list:
    if item[4] == '臺中市' and item[0] == '2021':
        taichung_2021.append(item)
print(taichung_2021)
```

```python
all = {'台北':taipei_2021,'台中':taichung_2021}
all
```
