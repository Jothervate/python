```python
import random as rd
students = []
for _ in range(50):
    students.append([rd.randint(50,100) for _ in range(5)])
students
```

```python
print("國文\t英文\t數學\t社會\t自然\t總分\t平均")
for student in students:
    sum = 0
    for item in student:
        sum += item
        print(item,end="\t")
    print(sum, end="\t")
    print(sum/5)
    
```
