```python
#讀取csv格式檔案
import csv

with open('歷年國內主要觀光遊憩據點遊客人數月別統計.csv',encoding="utf-8",newline="") as file:
    csv_reader = csv.reader(file)
    all_list = list(csv_reader)
all_list[-1:]
```

```python
taipei_2021 = []
for item in all_list:
    if item[4] == '臺北市' and item[0] == '2021':
        taipei_2021.append(item)
print(taipei_2021)
```

```python
taichung_2021 = []
for item in all_list:
    if item[4] == '臺中市' and item[0] == '2021':
        taichung_2021.append(item)
print(taichung_2021)
```

```python
import xlwt
wb = xlwt.Workbook()
ws = wb.add_sheet('A Test Sheet')

ws.write(0, 0, 1234.56)
ws.write(1, 0, 12345)
ws.write(2, 0, 1)
ws.write(2, 1, 1)
ws.write(2, 2, 1)

wb.save('example.xls')
```

```python
import xlwt
wb = xlwt.Workbook()
ws = wb.add_sheet('台北2021')
for rowIndex,row in enumerate(taipei_2021):    
    for columnIndex,column in enumerate(row):
        ws.write(rowIndex, columnIndex, column)
wb.save('台北.xls')
    
```
