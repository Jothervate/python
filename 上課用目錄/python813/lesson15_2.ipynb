```python
import numpy as np
import pandas as pd

scores = np.random.randint(50,101,size=(50,5))
scores
```

```python
class1 = pd.DataFrame(scores,
             index=[f'學生{v}' for v in range(1,51)],
             columns=['國文','英文','數學','自然','地理'])
class1
```

```python
sum = class1['國文'] + class1['英文'] +  class1['數學'] +  class1['自然'] + class1['地理']
sum
```

```python
class1['總分'] = sum
class1['平均'] = sum/5
class1['排名'] = sum.rank(method='first',ascending=False)
class1
```
