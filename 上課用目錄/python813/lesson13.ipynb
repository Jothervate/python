```python
import matplotlib.pyplot as plt

#index = list(range(5))
index = [3, 4, 5, 6, 7]
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
values = [5000, 6000, 7000, 8000, 9000]
figure = plt.figure()
axes = figure.add_subplot()
axes.bar(index, values)
axes.set_xticks(index,labels=subject)
for idx,i in enumerate(index):
    axes.annotate(str(values[idx]),(i-0.2,values[idx]+100))
#取得x,y軸最小,最大座標
xmin,xmax,ymin,ymax = axes.axis()
xmin -= 0.2
xmax += 0.2
ymin = 3000
ymax += 1000
axes.axis(xmin = xmin, xmax=xmax, ymin=ymin, ymax=ymax)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.font_manager import FontProperties

font=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=7)
font1=FontProperties(fname='/Library/Fonts/Arial Unicode.ttf',size=12)
#window font=FontProperties(fname=r'c:\windows\fonts\simsun.ttc',size=8)

#index = list(range(5))
index = np.array([3, 4, 5, 6, 7])
subject = ['apple', 'banana', 'orange', 'tomato', 'guava']
values1 = [5000, 6000, 7000, 8000, 9000]
values2 = [3000, 4000, 5000, 6000, 7000]
values3 = [4000, 5000, 6000, 7000, 8000]

figure = plt.figure()
axes = figure.add_subplot()
bin_width = 0.3
barContainer1 = axes.bar(index, values1,width=bin_width,color='red')
barContainer2 = axes.bar(index+bin_width, values2,width=bin_width,color='blue')
barContainer3 = axes.bar(index-bin_width, values3,width=bin_width,color='green')
axes.set_xticks(index,labels=subject)
#annotate
for idex,rec in enumerate(barContainer1.patches):
    x = rec.get_x()
    y = rec.get_y()    
    width = rec.get_width()
    height = rec.get_height()
    axes.annotate(str(values1[idex]),xy=(x+width/2-0.1,height+100),fontproperties=font)

for idex,rec in enumerate(barContainer2.patches):
    x = rec.get_x()
    y = rec.get_y()    
    width = rec.get_width()
    height = rec.get_height()
    axes.annotate(str(values2[idex]),xy=(x+width/2-0.1,height+100),fontproperties=font)

for idex,rec in enumerate(barContainer3.patches):
    x = rec.get_x()
    y = rec.get_y()    
    width = rec.get_width()
    height = rec.get_height()
    axes.annotate(str(values3[idex]),xy=(x+width/2-0.1,height+100),fontproperties=font)
    
#取得x,y軸最小,最大座標
xmin,xmax,ymin,ymax = axes.axis()
xmin -= 0.2
xmax += 0.2
ymin = 1000
ymax += 1000
axes.axis(xmin = xmin, xmax=xmax, ymin=ymin, ymax=ymax)
plt.show()
```
