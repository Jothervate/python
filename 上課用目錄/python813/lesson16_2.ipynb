```python
#各鄉鎮市區人口密度
#statistic_yyy(統計年)、site_id(區域別)、people_total(年底人口數)、area(土地面積)、population_density(人口密度)

url = "https://data.moi.gov.tw/MoiOD/System/DownloadFile.aspx?DATA=C45AD94C-ED23-4381-8F3E-359CD5C276F0"

#直接下載內容
import requests

response = requests.get(url)
if response.ok:
    print(response.text)
```

```python
import csv
import io
import pandas as pd

file = io.StringIO(response.text)
csvReader = csv.reader(file)
next(csvReader)
csv_list = list(csvReader)
dataFrame = pd.DataFrame(csv_list)
dataFrame
```

```python
dataFrame.columns = dataFrame.loc[0]
dataFrame
```

```python
dataFrame1 = dataFrame.drop(0)
dataFrame1
```

```python
#dataFrame1.iloc[:,1:4]
dataFrame2=dataFrame1.reindex(columns=['區域別','年底人口數','土地面積'])
dataFrame3 = dataFrame2.rename(columns={'年底人口數':'110年底人口數'})
dataFrame3
```
