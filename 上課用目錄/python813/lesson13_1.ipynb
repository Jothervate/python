```python
import matplotlib.pyplot as plt
labels = ['Nokia', 'Samsung', 'Apple', 'Lumia']
values = [20, 30, 45, 10]
colors =['yellow', 'green', 'red', 'blue']

figure = plt.figure()
axes = figure.add_subplot()
axes.pie(values,labels=labels,colors=colors,explode=[0.3,0,0,0],startangle=180,shadow=True,autopct='%1.1f%%')
plt.show()
```
