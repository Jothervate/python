```python
student1 = {"chinese":78, "english":80, "math":92, "history":75, "discover":85}
student2 = {"chinese":92, "english":60, "math":85, "history":62, "discover":73}

subject = list(student1.keys())
student1_num = list(student1.values())
student2_num = list(student2.values())
print(subject)
print(student1_num)
print(student2_num)
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(num="fig1",figsize=(8,4),dpi=220,facecolor='#cccccc')
ax = fig.add_subplot(1,1,1)
ax.plot(subject,student1_num,subject,student2_num)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(num="fig1",figsize=(8,4),dpi=220,facecolor='lightgray')
ax = fig.add_subplot(1,1,1)
ax.bar([0,1,2,3,4],student1_num,0.3)
ax.bar([0.3,1.3,2.3,3.3,4.3],student2_num,0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(num="fig1",figsize=(8,4),dpi=220,facecolor='lightgray')
ax = fig.add_subplot(1,1,1)
ax.barh([0,1,2,3,4],student1_num,0.3)
ax.barh([0.3,1.3,2.3,3.3,4.3],student2_num,0.3)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure(num="fig1",figsize=(8,4),dpi=220,facecolor='#cccccc')
ax = fig.add_subplot(1,1,1)
ax.plot(subject,student1_num,'m-.',subject,student2_num,'c-')
plt.show()
```

```python
#grid()
x = [1,2,3,4,5]
y1 = [1,2,5,6,7]
y2 = [4,5,6,7,8]
fig = plt.figure(num="fig1",figsize=(8,4),dpi=220,facecolor='#cccccc')
ax = fig.add_subplot(1,1,1)
ax.plot(x,y1,'r-o',x,y2,'g--.')
ax.grid(axis='y')
plt.show()
```

```python
#axis()
x = [1,2,3,4,5]
y1 = [1,2,5,6,7]
y2 = [4,5,6,7,8]
fig = plt.figure(num="fig1",figsize=(8,4),dpi=220,facecolor='#cccccc')
ax = fig.add_subplot(1,1,1)
ax.plot(x,y1,'r-o',x,y2,'g--.')
ax.grid(axis='y')
ax.axis(xmin=2.0, xmax=3.5, ymin=1.5, ymax=6.5)
plt.show()
```

```python
#set_title(),set_xlabe(),set_ylabel()
import random

days = [num for num in range(1,31)]
celsius = [round(random.uniform(28.5,35.2),1) for _ in range(1,31)]
fig = plt.figure(num="fig1",figsize=(8,4),dpi=220,facecolor='#cccccc')
ax = fig.add_subplot(1,1,1)
ax.plot(days,celsius,'ro')
ax.grid(axis='x')
ax.set_title("july weather")
ax.set_xlabel("Days")
ax.set_ylabel("Celsius")
plt.show()
```
