```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax_1 = fig.add_subplot(1,1,1)
x = [0, 1, 2]
y = [3, 4, 5]
ax_1.plot(x,y,'bo')
print(type(ax_1))
ax_1.set_title("First Line Plot")
plt.show()
```
